# US Exports and Imports
---


```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
file = "imports_usafacts.csv"
df = pd.read_csv(file)
df.head()
```

Lets first remove the last five rows since it just has some nonuseful metadata

```python
df = df.iloc[:-5,:]
```

Now lets transpose the dataframe so that the variable of intrest are in the columns and not in the rows

```python
df = df.transpose()
```

The column names are in the first row, so lets rename the columns to that and remove the first row

```python
df.columns = df.iloc[0]
df.head()
```

```python
df = df.iloc[1:]
df.head()
```

Looking at the data, we are only intrested in the export of goods. Therefore we will remove the last columns that are related to services and investments 

```python
df = df.iloc[:,:49]
```

Lets only take a look at the last 10 years in the dataset

```python
df = df.assign(Year = lambda x: pd.to_datetime(x.index)).loc['2008':]
df.head()
```

Finally lets rename the columns to make subsetting easier

```python
df.columns = df.columns.str.replace('($)','',regex=False).str.strip()
df.head()
```

# EDA
   
Let's take a look at some graphs.   

First let's see a time series of the exports from Mexico and Canada to the US and observe any trends.

```python
fig,ax = plt.subplots(figsize=(12,6))
(df[['Canada','Mexico']] / 1e9).plot(ax=ax,marker='o',linewidth=3,markersize=10)
ax.set_xlabel('Year')
ax.set_ylabel('Exports to US (Billion $)')
ax.set_title('Trade Between NAFTA Nations')
plt.show()
```

First off we can see a big dip in both countries following 2008. This is following the great recession, which may account for the dip.   

We can see that Canada had a major downshift in exports to the US following 2014. Doing some [research](https://www.wsj.com/articles/canada-economic-growth-slowed-at-end-of-2015-1456846514) reveals this is around the time that crude oil prices dropped dramatically which heavily impacted the oil industry of Canada. Mexico imports slowed down during this period, but managed to maintain and eventually overtake Canada.

```python
fig,ax = plt.subplots(figsize=(12,6))
(df[['Mexico','Imports']] / 1e9).plot.scatter(ax=ax,x='Imports',y='Mexico',figsize=(12,6),s=100)
ax.set_xlabel('Total US Imports (Billion $)')
ax.set_ylabel('Mexico Exports to US (Billion $)')
ax.set_title('Mexico Exports to US Correlated with Total US Imports')
plt.show()
```
